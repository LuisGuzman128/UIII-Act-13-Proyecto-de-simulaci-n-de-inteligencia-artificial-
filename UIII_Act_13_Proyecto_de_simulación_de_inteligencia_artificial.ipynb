```python
# 22/mayo/2026

# Cuenta_Corriente:
# Número de cuenta, nombre del cliente,
# saldo actual, tipo de cuenta,
# límite de retiro y estado financiero.

# ======================================================
# 💳 CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Manuel Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 27
# 💳 Proyecto: Simulación de Inteligencia Artificial
# ======================================================

# ======================================================
# DEFINICIÓN DE ATRIBUTOS
# ======================================================

# numero_cuenta:
# Guarda el número único de la cuenta bancaria.

# nombre_cliente:
# Guarda el nombre del propietario de la cuenta.

# saldo_actual:
# Guarda el dinero disponible en la cuenta.

# tipo_cuenta:
# Guarda el tipo de cuenta bancaria
# (ahorro, nómina o empresarial).

# limite_retiro:
# Guarda el monto máximo permitido para retiro.

# ======================================================
# DEFINICIÓN DE FUNCIONES
# ======================================================

# agregar_cuenta():
# Permite registrar una nueva cuenta corriente.

# listar_cuentas():
# Muestra todas las cuentas en forma de tabla.

# actualizar_cuenta():
# Permite modificar los datos de una cuenta.

# borrar_cuenta():
# Elimina una cuenta registrada.

# analizar_cuenta():
# Analiza el estado financiero del cliente.

# ======================================================
# CLASE PRINCIPAL
# ======================================================

class Cuenta_Corriente:

    def __init__(self):

        self.cuentas = {}

    # ==================================================
    # AGREGAR CUENTA
    # ==================================================

    def agregar_cuenta(self):

        numero = input("Ingresa número de cuenta: ")
        nombre = input("Ingresa nombre del cliente: ")
        saldo = float(input("Ingresa saldo actual: "))
        tipo = input("Ingresa tipo de cuenta: ")
        limite = float(input("Ingresa límite de retiro: "))

        self.cuentas[numero] = {

            "Nombre": nombre,
            "Saldo": saldo,
            "Tipo": tipo,
            "Limite": limite
        }

        print("\n✅ Cuenta agregada correctamente")

    # ==================================================
    # LISTAR CUENTAS
    # ==================================================

    def listar_cuentas(self):

        print("\n================================================================================================================")
        print(f"{'CUENTA':<15}{'CLIENTE':<25}{'SALDO':<15}{'TIPO':<20}{'LIMITE RETIRO':<20}")
        print("================================================================================================================")

        for key, value in self.cuentas.items():

            print(f"{key:<15}"
                  f"{value['Nombre']:<25}"
                  f"{value['Saldo']:<15}"
                  f"{value['Tipo']:<20}"
                  f"{value['Limite']:<20}")

        print("================================================================================================================")

    # ==================================================
    # ACTUALIZAR CUENTA
    # ==================================================

    def actualizar_cuenta(self):

        cuenta = input("Ingresa número de cuenta a actualizar: ")

        if cuenta in self.cuentas:

            nombre = input("Nuevo nombre del cliente: ")
            saldo = float(input("Nuevo saldo actual: "))
            tipo = input("Nuevo tipo de cuenta: ")
            limite = float(input("Nuevo límite de retiro: "))

            self.cuentas[cuenta] = {

                "Nombre": nombre,
                "Saldo": saldo,
                "Tipo": tipo,
                "Limite": limite
            }

            print("\n✅ Cuenta actualizada correctamente")

        else:
            print("❌ Cuenta no encontrada")

    # ==================================================
    # BORRAR CUENTA
    # ==================================================

    def borrar_cuenta(self):

        cuenta = input("Ingresa número de cuenta a borrar: ")

        if cuenta in self.cuentas:

            del self.cuentas[cuenta]

            print("\n✅ Cuenta eliminada correctamente")

        else:
            print("❌ Cuenta no encontrada")

    # ==================================================
    # ANALIZAR CUENTA
    # ==================================================

    def analizar_cuenta(self):

        cuenta = input("Ingresa número de cuenta a analizar: ")

        if cuenta in self.cuentas:

            saldo = self.cuentas[cuenta]["Saldo"]

            print("\n=========== ANALISIS IA ===========")

            if saldo >= 50000:
                print("💎 Cliente Premium")

            elif saldo >= 10000:
                print("💰 Cliente Estable")

            elif saldo >= 1000:
                print("📈 Cliente Regular")

            else:
                print("⚠️ Cliente con saldo bajo")

        else:
            print("❌ Cuenta no encontrada")

# ======================================================
# OBJETO PRINCIPAL
# ======================================================

sistema = Cuenta_Corriente()

# ======================================================
# MENU PRINCIPAL
# ======================================================

while True:

    print("\n=========== MENU CUENTA CORRIENTE 💳 ===========")
    print("1. Agregar cuenta")
    print("2. Listar cuentas")
    print("3. Actualizar cuenta")
    print("4. Borrar cuenta")
    print("5. Analizar cuenta")
    print("6. Salir")

    opcion = int(input("Selecciona una opción: "))

    if opcion == 1:
        sistema.agregar_cuenta()

    elif opcion == 2:
        sistema.listar_cuentas()

    elif opcion == 3:
        sistema.actualizar_cuenta()

    elif opcion == 4:
        sistema.borrar_cuenta()

    elif opcion == 5:
        sistema.analizar_cuenta()

    elif opcion == 6:

        print("\n💳 Programa finalizado")
        break

    else:
        print("❌ Opción no válida")
```
